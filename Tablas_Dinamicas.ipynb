### NUEVA DATA

```python
data = {
    'ciudad': ['Bogota', 'Bogota', 'Medellin', 'Cali', 'Cali', 'Medellin'],
    'producto': ['Laptop', 'Celular', 'Laptop', 'Celular', 'Laptop', 'Celular'],
    'cantidad': [10, 5, 8, 3, 2, 7]
}
ventas_dataframe = pd.DataFrame(data)

agrupacion_ventas = ventas_dataframe.pivot_table(
    index="ciudad",
    columns="producto",
    values="cantidad",
    aggfunc=sum,
    fill_value=0
)

agrupacion_ventas
    
```
