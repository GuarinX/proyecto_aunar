## Variables Categóricas

* **Gráfico de barras**

> Un gráfico de barras es una representación visual de datos donde las categorías se muestran en el eje horizontal (x) y los valores correspondientes en el eje vertical (y) mediante barras rectangulares. La longitud o altura de cada barra es proporcional al valor que representa. Se utiliza para comparar diferentes categorías o mostrar cambios en datos a lo largo del tiempo.

</div>

* **Gráfico circular**

> Un gráfico circular es un diagrama en forma de círculo dividido en sectores, donde cada sector representa una proporción del total. La magnitud de cada sector es proporcional a la cantidad o porcentaje que representa. Se utiliza para mostrar la distribución de partes dentro de un todo y es ideal para comparar proporciones o porcentajes.

## Variables Numéricas

1. **Una Variable**
    * `Histograma`

> Un histograma es un tipo de gráfico que muestra la distribución de un conjunto de datos numéricos mediante barras. Cada barra representa un rango de valores (intervalo) y su altura indica la cantidad de datos que caen dentro de ese rango.

<div style="text-align: center;">
    <img src="media/histograma.png" alt="Histograma" width="50%">
</div>

2. **Dos Variables**
    * `Boxplot`

> Un boxplot es un gráfico que resume la distribución de un conjunto de datos numéricos utilizando cinco medidas principales: el valor mínimo, el primer cuartil (Q1), la mediana, el tercer cuartil (Q3) y el valor máximo. El gráfico consiste en una "caja" que abarca el rango intercuartílico (de Q1 a Q3) y una línea dentro de la caja que indica la mediana.

<div style="text-align: center;">
    <img src="media/boxplot.png" alt="Boxplot" width="50%">
</div>

* `Scatterplot`

> Un scatterplot es un gráfico que utiliza puntos para representar los valores de dos variables numéricas en un plano cartesiano. Cada punto en el gráfico indica un valor para una variable en el eje horizontal (x) y un valor para otra variable en el eje vertical (y).

<div style="text-align: center;">
    <img src="media/scatter.png" alt="Scatterplot" width="50%">
</div>


```python
#impotrtamos la libreria matplotlib
import matplotlib.pyplot as plt
```

### CREAMOS LOS DATOS

```python
labels = ['Ingenieria Informatica', 'Administración de empresas', 'Diseño Grafico', 'Seguridad y Salud']

# valores de cada categoría
sizes = [30, 45, 15, 30]

# Colores para cada sección
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  
```

### CREAMOS EL DIAFRAMA CIRCULAR

```python
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('DIAGRAMA CIRCULAR')
plt.show()
```
